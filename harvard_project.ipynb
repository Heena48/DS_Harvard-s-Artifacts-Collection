```python
import sqlite3
import pandas as pd


conn = sqlite3.connect(r"C:\Users\Windows\Downloads\HarvardMuseum.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS artifact_colors (
    objectid INTEGER,
    spectrum TEXT,
    hue TEXT,
    percent REAL,
    css3 TEXT
);
""")
conn.commit()
cursor.execute("""
CREATE TABLE IF NOT EXISTS artifact_metadata (
    id INTEGER PRIMARY KEY,
    title TEXT,
    culture TEXT,
    period TEXT,
    century TEXT,
    medium TEXT,
    dimensions TEXT,
    department TEXT,
    classification TEXT,
    accessionyear INTEGER,
    accessionmethod TEXT
);
""")
conn.commit()
cursor.execute("""
CREATE TABLE IF NOT EXISTS artifact_media (
    objectid INTEGER,
    imagecount INTEGER,
    mediacount INTEGER,
    colorcount INTEGER,
    rank INTEGER,
    datebegin INTEGER,
    dateend INTEGER,
    FOREIGN KEY (objectid) REFERENCES artifact_metadata(id)
);
""")
conn.commit()

cursor.execute("""
DELETE FROM artifact_colors;
""")
conn.commit()
cursor.execute("""
DELETE FROM artifact_metadata;
""")
conn.commit()
cursor.execute("""
DELETE FROM artifact_media;
""")
conn.commit()

df_colors.to_sql('artifact_colors', conn, if_exists='append', index=False)
df_metadata.to_sql('artifact_metadata', conn, if_exists='append', index=False)
df_mediadata.to_sql('artifact_media', conn, if_exists='append', index=False)
conn.close()
```
